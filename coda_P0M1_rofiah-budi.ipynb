Program ini dibuat untuk melakukan pengambilan data dari halaman website footlocker menggunakan library selenium, kemudian diolah dan disimpan ke dalam database PostgreSQL.

```python
# data manipulation or dataframe
import pandas as pd
import time

# webscraping
from selenium import webdriver
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

Import semua library yang akan digunakan untuk data manipulation dan webscrapping menggunakan selenium

```python
# empty list
names = []
brands = []
prices = []
categories = []
colors = []
boxes = []

# create driver
driver = webdriver.Chrome()

for page in range(1, 4):
    url = f'https://www.footlocker.id/catalogsearch/result/?p={page}&q=shoes'
    driver.get(url)
    time.sleep(10)

    # fetch html components
    html = driver.page_source

    # parse
    soup = BeautifulSoup(html, "html.parser")

    boxes = soup.find_all('li', {'class':'product-card item product product-item'})

    # <a class="product-item-link" href="https://www.footlocker.id/speedcat-og-unisex-lifestyle-shoes-black-16.html" data-item-page="1">Speedcat OG Unisex Lifestyle Shoes - Black                </a>
    #
    for box in boxes:
        try:
            el_name = box.find('a', {'class':'product-item-link'})
            names.append(el_name.get_text())
        except:
            names.append(None)

        try:
            el_brand = box.find('h3', {'class':'product-brand'})
            brands.append(el_brand.get_text())
        except:
            names.append(None)
        
        try:
            el_price = box.find('span', {'data-price-type':'finalPrice'})
            prices.append(el_price.get_text())
        except:
            prices.append(None)

        try:
            el_categorie = box.find('div', {'class':'gender'})
            categories.append(el_categorie.get_text())
        except:
            prices.append(None)

        try:
            el_color = box.find('span', {'class':'value'})
            colors.append(el_color.get_text())
        except:
            colors.append(None)

driver.quit()
```

Melakukan scrapping dengan menggunakan library selenium, dengan method halaman page yaitu looping untuk setiap pagenya, disini saya mengambil 3 page karena dirasa sudah cukup untuk mendapat lebih dari 50 data dengan mengatur time.sleep ke 10s, dengan get.element yaitu nama produk, merk, harga, kategori, dan juga warna (5 columns)
Menggunakan method try except karena apabila dalam 1 row terdapat element yang tidak ada maka akan di append sebagai NaN

```python
# create a dataframe
df = pd.DataFrame(
    {
        'nama produk':names,
        'merk':brands,
        'harga':prices,
        'kategori':categories,
        'warna':colors
    }
)
```
